### Create Graphs using LangChain and then export using Neo4j Bloom in *.json
Access tool using: [Graph Builder](https://llm-graph-builder.neo4jlabs.com/)

```python
import json
import networkx as nx
```

```python
# Step 1: Read the JSON file
with open('./Output/neo4j_query_table_data_2024-11-15.json', 'r') as file:
    data = json.load(file)

# Step 2: Parse the JSON data
# Assuming the JSON structure has nodes and edges
nodes = []
edges = []

# Example of extracting nodes and edges from the JSON structure
# This part needs to be adapted based on the actual structure of your JSON
for element in data:
    # if 'start' in element and 'end' in element:        
        start_node = element['p']['start']['identity']
        end_node = element['p']['end']['identity']
        nodes.append(start_node)
        nodes.append(end_node)
        edges.append((start_node, end_node))

# Step 3: Create a NetworkX graph
G = nx.Graph()

# Add nodes and edges to the graph
G.add_nodes_from(nodes)
G.add_edges_from(edges)

# Now you can work with the graph G
print("Nodes:", G.nodes())
print("Edges:", G.edges())
```

```python
# Print graph
nx.draw(G,
        with_labels=True, 
        node_size=0.1, 
        node_color="skyblue", 
        node_shape=".", 
        alpha=1, 
        linewidths=60)
```
